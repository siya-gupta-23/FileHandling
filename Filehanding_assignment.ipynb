```python
'''4. Write a Python program using multithreading where one thread adds numbers to a list
     and another thread removes numbers from the list. Implement a mechanism to avoid race conditions using threading.Lock.'''


import threading
import time

# Shared resource
data_list = []
data_lock = threading.Lock()

# Function to add numbers to the list
def add_to_list():
    for number in range(10):
        time.sleep(1)  
        with data_lock:
            data_list.append(number)
            print(f"Added {number} to the list.")
        time.sleep(1)

# Function to remove numbers from the list
def remove_from_list():
    for _ in range(10):
        time.sleep(2)  
        with data_lock:
            if data_list:
                removed_value = data_list.pop(0)
                print(f"Removed {removed_value} from the list.")
            else:
                print("List is empty, nothing to remove.")
        time.sleep(2)

# Create threads
add_thread = threading.Thread(target=add_to_list)
remove_thread = threading.Thread(target=remove_from_list)

# Start threads
add_thread.start()
remove_thread.start()

# Wait for threads to complete
add_thread.join()
remove_thread.join()

print("Final state of the list:", data_list)
```

```python
#7. Create a program that uses a thread pool to calculate the factorial of numbers from 1 to 10 concurrently.Use concurrent.futures.ThreadPoolExecutor to manage the threads.

import concurrent.futures
import math

# calculate factorial
def calculate_factorial(n):
    print(f"Calculating factorial of {n}")
    return math.factorial(n)

# Main block to use ThreadPoolExecutor
if __name__ == '__main__':
    numbers = list(range(1, 11))  # Numbers from 1 to 10
    futures = []

    # Create a thread pool with ThreadPoolExecutor
    with concurrent.futures.ThreadPoolExecutor() as executor:
        # pass each number to calculate_factorial function
        for number in numbers:
            future = executor.submit(calculate_factorial, number)
            futures.append(future)

        # when  the futures complete it will print the results
        for future in concurrent.futures.as_completed(futures):
            result = future.result()
            print(f"Factorial calculated: {result}")
```

```python
'''8. Create a Python program that uses multiprocessing.Pool to compute the square of numbers from 1 to 10 in
parallel. Measure the time taken to perform this computation using a pool of different sizes (e.g., 2, 4, 8
processes).'''

import multiprocessing
import time

# calculate the square of a number
def compute_square(n):
    return n * n

# Main function to calculate squares using multiprocessing Pool with map()
def compute_squares_with_pool(pool_size):
    numbers = list(range(1, 11))  # Numbers from 1 to 10

    # Start time measurement
    start_time = time.time()

    # Create a multiprocessing pool with 'pool_size' processes
    with multiprocessing.Pool(pool_size) as pool:
        # Using map() to distribute tasks
        results = pool.map(compute_square, numbers)

    # End time measurement
    end_time = time.time()

    # Print the results and the time taken
    print(f"Results with {pool_size} processes: {results}")
    print(f"Time taken with {pool_size} processes: {end_time - start_time:.4f} seconds\n")

if __name__ == '__main__':
    # Test with different pool sizes (2, 4, and 8)
    for pool_size in [2, 4, 8]:
        compute_squares_with_pool(pool_size)

# i have used Ai tool to understand the questions and how it will be implement cause lack of understanding of this topic.
```
